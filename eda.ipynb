```python
df['Status'].value_counts().plot(kind='bar', color=['#FF9999', '#66B3FF'])
plt.title('Loan Status Distribution')
```

```python
# Find and display the count of null values in each column
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)

```

```python
# Fill null values with appropriate values
# For numerical columns, fill with the mean
# For categorical columns, fill with the mode
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)


print("Null values after filling:\n", df.isnull().sum())
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
# Compute correlation matrix (numeric features)
corr = numeric_df.corr()
#correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix (`corr`) provides insights into the relationships between numerical variables in the dataset. Here's what it shows:

1. **Strong Positive Correlations**:
    - `loan_amount` and `property_value` (0.688): Indicates that higher loan amounts are associated with higher property values.
    - `rate_of_interest` and `Interest_rate_spread` (0.614): Suggests that as the interest rate increases, the spread also tends to increase.

2. **Moderate Positive Correlations**:
    - `term` and `dtir1` (0.104): Indicates a slight positive relationship between loan term and debt-to-income ratio.
    - `LTV` and `dtir1` (0.072): Suggests a weak positive relationship between loan-to-value ratio and debt-to-income ratio.

3. **Negative Correlations**:
    - `property_value` and `LTV` (-0.215): Indicates that higher property values are associated with lower loan-to-value ratios.
    - `income` and `dtir1` (-0.253): Suggests that higher incomes are associated with lower debt-to-income ratios.

4. **Weak or No Correlation**:
    - Many variables, such as `Credit_Score` and `loan_amount`, show very weak or no significant correlation with other variables.

### Key Takeaways:
- The matrix helps identify relationships that could be useful for predictive modeling or understanding trends in the data.
- Strong correlations (positive or negative) can indicate potential dependencies or patterns between variables.
- Weak correlations suggest little to no linear relationship between the variables.
