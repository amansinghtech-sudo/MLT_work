```python
import pandas as pd

#loading the full dataset of around 1.2 GB
full_df = pd.read_csv('loan.csv')
```

```python
from sklearn.model_selection import train_test_split

sample_df,_ = train_test_split(full_df, test_size=0.98,stratify= full_df['loan_status'], random_state=42) #keep around 2% of the data
#saving the sample dataset to a csv file
sample_df.to_csv('sample_loan.csv', index=False)
```
